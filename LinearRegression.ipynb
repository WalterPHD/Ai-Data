```python
import numpy as np

class ScratchLinearRegression():
    def __init__(self, num_iter=1000, lr=0.01, bias=True, verbose=False):
        self.iter = num_iter
        self.lr = lr
        self.bias = bias
        self.verbose = verbose
        # Initialize loss for training data
        self.loss = np.zeros(self.iter)
        # Initialize loss for validation data (add this attribute as it's used later)
        self.val_loss = np.zeros(self.iter)

    # Corrected fit function signature to include validation data
    def fit(self, X, y, X_val=None, y_val=None):
        m, n = X.shape

        # Add bias column if bias is enabled
        if self.bias:
            X = np.c_[np.ones((m, 1)), X]
            # Add bias column to validation data if provided
            if X_val is not None:
                X_val = np.c_[np.ones((X_val.shape[0], 1)), X_val]

        # Initialize weights (coefficients)
        self.coef_ = np.zeros(X.shape[1])

        # Training loop
        for i in range(self.iter):
            # Calculate predictions for training data
            y_pred = X.dot(self.coef_)
            # Calculate error for training data
            error = y_pred - y

            # Calculate gradient
            gradient = (1 / m) * X.T.dot(error)
            # Update weights
            self.coef_ -= self.lr * gradient

            # Calculate training loss (Mean Squared Error)
            J = (1 / (2 * m)) * np.sum(error ** 2)
            self.loss[i] = J

            # Calculate validation loss if validation data is provided
            if X_val is not None and y_val is not None:
                y_val_pred = X_val.dot(self.coef_)
                val_error = y_val_pred - y_val
                J_val = (1 / (2 * X_val.shape[0])) * np.sum(val_error ** 2)
                self.val_loss[i] = J_val

            # Print verbose output
            if self.verbose:
                if X_val is not None and y_val is not None:
                    print(f"Iteração {i+1}: Loss treino = {J:.6f}, Loss validação = {J_val:.6f}")
                else:
                    print(f"Iteração {i+1}: Loss treino = {J:.6f}")

    def predict(self, X):
        # Add bias column to input data for prediction if bias is enabled
        if self.bias:
            X = np.c_[np.ones((X.shape[0], 1)), X]
        # Calculate predictions
        return X.dot(self.coef_)

    def mean_squared_error(self, X, y):
        # Calculate predictions
        y_pred = self.predict(X)
        # Calculate Mean Squared Error
        mse = np.mean((y - y_pred) ** 2)
        return mse
```

# Step 1: Define the Objective Function (Cost Function)

We want to minimize the Mean Squared Error (MSE), defined as:

$$
J(\theta) = \frac{1}{2m} \sum_{i=1}^m \left( h_\theta(x^{(i)}) - y^{(i)} \right)^2
$$

* $m$: number of samples
* $h_\theta(x^{(i)}) = \theta^T x^{(i)}$: hypothesis or prediction for the $i$-th sample
* $y^{(i)}$: true label of the $i$-th sample

The factor $\frac{1}{2m}$ is used to simplify the gradient calculation later.

---

# Step 2: Update Parameters Using Gradient Descent

We update each parameter $\theta_j$ by moving it opposite to the gradient of the cost function:

$$
\theta_j := \theta_j - \alpha \frac{\partial}{\partial \theta_j} J(\theta)
$$

* $\alpha$: learning rate
* $\frac{\partial}{\partial \theta_j} J(\theta)$: gradient of the cost with respect to $\theta_j$

---

# Step 3: Compute Partial Derivative of the Cost Function w\.r.t. $\theta_j$

Calculate:

$$
\frac{\partial}{\partial \theta_j} J(\theta) = \frac{\partial}{\partial \theta_j} \left( \frac{1}{2m} \sum_{i=1}^m (h_\theta(x^{(i)}) - y^{(i)})^2 \right)
$$

Using the chain rule and the fact that

$$
h_\theta(x^{(i)}) = \sum_{k=0}^n \theta_k x_k^{(i)}
$$

we get:

$$
\frac{\partial}{\partial \theta_j} J(\theta) = \frac{1}{m} \sum_{i=1}^m \left( h_\theta(x^{(i)}) - y^{(i)} \right) x_j^{(i)}
$$

---

# Step 4: Final Update Formula

Plugging the gradient back into the gradient descent update, we have:

$$
\boxed{
\theta_j := \theta_j - \alpha \frac{1}{m} \sum_{i=1}^m \left( h_\theta(x^{(i)}) - y^{(i)} \right) x_j^{(i)}
}
$$

---
